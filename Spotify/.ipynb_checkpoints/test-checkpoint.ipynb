```python
#Read last fifty songs
query = "/me/player/recently-played"
params = {"limit": 50}
header = {"Authorization": "Bearer {}".format(token)}
response = requests.get("{}{}".format(base_query, query),
                        params = params,
                        headers = header
                       )
response
```

```python
dat = response.json()
```

```python
dat["items"][0]["track"]["artists"]
```

```python
#History table
timestamp = []
name = []
spot_id = []
uri = []
pop = []
types = []

artist = []
album = []

for element in dat["items"]:
    timestamp.append(element["played_at"])
    name.append(element["track"]["name"])
    spot_id.append(element["track"]["id"])
    uri.append(element["track"]["uri"])
    types.append(element["context"]["type"])
    pop.append(element["track"]["popularity"])
    artist.append(",".join([artist["name"] for artist in element["track"]["artists"]]))
    album.append(element["track"]["album"]["name"])
                  
data = pd.DataFrame({"Timestamp":timestamp,
                     "Track Name":name,
                     "Spotify ID": spot_id,
                     "Spotify URI": uri,
                     "Popularity": pop,
                     "Object Type": types
                    })
```

```python
query = "https://api.spotify.com/v1/audio-features"

response = requests.get(query, params = {"ids": ",".join(spot_id)}, headers = {"Authorization": "Bearer {}".format(token)})
```

```python
audio_feat = response.json()
```

```python
keys = ["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", "liveness", 
      "valence", "tempo", "duration_ms", "time_signature", "id", "uri"]
data_dic = {}

for key in keys:
    data_dic[key] = []

for element in audio_feat["audio_features"]:
    for key in keys:
        data_dic[key].append(element[key])

test = pd.DataFrame(data_dic)
```
